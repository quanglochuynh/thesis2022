```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from image_extractor import preprocess_hsv
from skimage.feature import graycomatrix, graycoprops

classes_name = ['Agglutinated', 'Brittle', 'Compartmentalized_Brown', 'Compartmentalized_PartiallyPurple', 'Compartmentalized_Purple', 'Compartmentalized_Slaty', 'Compartmentalized_White', 'Flattened', 'Moldered', 'Plated_Brown', 'Plated_PartiallyPurple', 'Plated_Purple', 'Plated_Slaty', 'Plated_White']

image_dir  = 'D:/Thesis_data/Backups/Color_Corrected_512x512/'

level = 32
bins = np.linspace(0, 255+1, level+1)
        
```

```python
for i in range(len(classes_name)):
    cons_r = []
    disi_r = []
    corr_r = []
    asm_r = []
    homo_r = []
    cons_g = []
    disi_g = []
    corr_g = []
    asm_g = []
    homo_g = []
    cons_b = []
    disi_b = []
    corr_b = []
    asm_b = []
    homo_b = []
    for k in range(1,101):
        add = image_dir + classes_name[i] + 'image (' + str(k) + ').JPG'
        image  = cv2.imread(image_dir + classes_name[i] + '/image (' + str(k) + ').JPG')
        print(add)
        image_rgb = cv2.cvtColor(preprocess_hsv(image, Contour=False), cv2.COLOR_HSV2RGB)

        digitize = np.digitize(image_rgb[:,:,0], bins) - 1
        glcm = graycomatrix(digitize, [3,5,7,9,11], [0, np.pi/4, np.pi/2, 3*np.pi/4], level, True, True)
        glcm = glcm[1:level,1:level]
        cons_r.append(graycoprops(glcm, 'contrast'))
        disi_r.append(graycoprops(glcm, 'dissimilarity'))
        asm_r.append(graycoprops(glcm, 'ASM'))
        corr_r.append(graycoprops(glcm, 'correlation'))
        homo_r.append(graycoprops(glcm, 'homogeneity'))

        digitize = np.digitize(image_rgb[:,:,1], bins) - 1
        glcm = graycomatrix(digitize, [3,5,7,9,11], [0, np.pi/4, np.pi/2, 3*np.pi/4], level, True, True)
        glcm = glcm[1:level,1:level]
        cons_g.append(graycoprops(glcm, 'contrast'))
        disi_g.append(graycoprops(glcm, 'dissimilarity'))
        asm_g.append(graycoprops(glcm, 'ASM'))
        corr_g.append(graycoprops(glcm, 'correlation'))
        homo_g.append(graycoprops(glcm, 'homogeneity'))

        digitize = np.digitize(image_rgb[:,:,2], bins) - 1
        glcm = graycomatrix(digitize, [3,5,7,9,11], [0, np.pi/4, np.pi/2, 3*np.pi/4], level, True, True)
        glcm = glcm[1:level,1:level]
        cons_b.append(graycoprops(glcm, 'contrast'))
        disi_b.append(graycoprops(glcm, 'dissimilarity'))
        asm_b.append(graycoprops(glcm, 'ASM'))
        corr_b.append(graycoprops(glcm, 'correlation'))
        homo_b.append(graycoprops(glcm, 'homogeneity'))

    np.savez_compressed('D:/Thesis_data/mlp_data/glcm/glcm_feature_' + classes_name[i], cons=[cons_r, cons_g, cons_b], disi=[disi_r, disi_g, disi_b], asm=[asm_r, asm_g, asm_b], corr=[corr_r, corr_g, corr_b], homo=[homo_r, homo_g, homo_b])
```
