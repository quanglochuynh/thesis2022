```python
extractor = feature_extract()
image_bgr = cv2.imread(image_dir)
extractor.pre_process(image_bgr)
image_rgb = extractor.image_rgb
image_hsv = extractor.image_hsv
plt.imshow(image_rgb)
plt.show()
```

```python
gau = mh.gaussian_filter(image_hsv[:,:,2], 1)
plt.imshow(gau)
plt.show()
```

```python
threshed = (gau > gau.mean())
plt.imshow(np.asarray(threshed*255, dtype=np.uint8))
plt.show()
# print(gau)
```

```python
labeled, n = mh.label(threshed)
print("Labelled Image")
plt.imshow(labeled)
plt.show()
```

```python
h_feature = mh.features.haralick(image_hsv[:,:,2], compute_14th_feature=True)
 
print("Haralick Features")
plt.imshow(h_feature)
plt.show()
```
