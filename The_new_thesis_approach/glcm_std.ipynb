```python
import numpy as np
import cupy as cp
import matplotlib.pyplot as plt

classes_name = ['Agglutinated', 'Brittle', 'Compartmentalized_Brown', 'Compartmentalized_PartiallyPurple', 'Compartmentalized_Purple', 'Compartmentalized_Slaty', 'Compartmentalized_White', 'Flattened', 'Moldered', 'Plated_Brown', 'Plated_PartiallyPurple', 'Plated_Purple', 'Plated_Slaty', 'Plated_White']

dr = 'D:/Thesis_data/mlp_data/glcm/glcm_feature_'
```

```python
plt.rcParams["figure.figsize"] = (18,96)

for i in range(len(classes_name)):
    class_glcm = cp.load(dr + classes_name[3]+ '.npz', allow_pickle=True)
    cons = class_glcm['disi']
    std_cons_r = cp.asnumpy(cp.std(cons[0], axis=0))
    std_cons_g = cp.asnumpy(cp.std(cons[1], axis=0))
    std_cons_b = cp.asnumpy(cp.std(cons[2], axis=0))
    plt.subplot(14,3,i*3+1)
    plt.imshow(std_cons_r)
    plt.subplot(14,3,i*3+2)
    plt.imshow(std_cons_g)    
    plt.subplot(14,3,i*3+3)
    plt.imshow(std_cons_b)    
plt.show()
```
