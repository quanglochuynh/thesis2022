```python
import numpy as np
# from tensorflow import keras
import matplotlib.pylab as plt
import cv2
import os 
import pickle

data_dir = "D:/Thesis_data/testing_img"
categories = ["Agglutinated", "Brittle", "Compartmentalized_Brown", "Compartmentalized_PartiallyPurple", "Compartmentalized_Purple", "Compartmentalized_Slaty", "Compartmentalized_White", "Flattened", "Moldered", "Plated_Brown", "Plated_PartiallyPurple", "Plated_Purple", "Plated_Slaty", "Plated_White"]
```

```python
# x_train = np.array([],dtype="float32")
# y_train = np.array([],dtype="float32")
x_train = []
y_train = []
m=0
for i in categories:
    # print(m)
    path = os.path.join(data_dir, i)
    n = 0
    for img in os.listdir(path):
        if "image" in img:
            img = cv2.imread(os.path.join(path,img))
            # print(np.shape(img))
            # img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            x_train.append(img)
            # y_train.append(m)
            n = n+1
        # break
    # break
    m = m+1
# plt.show()
# y_train = keras.utils.to_categorical(y_train, 14)
# print(y_train)
print(np.shape(x_train))
# print(np.shape(y_train))
```

```python
filename = "D:/Thesis_data/x_test.plk"
outfile = open(filename, 'wb')
pickle.dump(x_train, outfile)

```
